#Formatting/Cleaning Data

We recall that our data had a lot of missing values for age. We should delete these values, as using the average age would instead be more problamatic for our conclusion, since our conclusion is based entirely on the frequency of certain age values, and ages are likely missing completely at random (MCAR).

Firstly, we need to reformat the age values. The age is additionally unusually formatted. There is a :: before each value of age. We can simply use a python string formatter and add all these values to a list. We'll do string slicing for this by just taking the last two values of age.

```python
dropped_gun = gun_violence.dropna(subset = ['participant_age'])
print("There are {0:,} rows of data after dropping, compared to {1:,} before.".format(len(dropped_gun),
                                                                                     len(gun_violence)))
```

We lost a lot of values, however this is still enough to make an accurate conclusion.

The age is additionally unusually formatted. There is a :: before each value of age. We can simply use a python string formatter and add all these values to a list. We'll do string slicing for this by just taking the last two values of age.

We'll convert all values in that column so we can use the slice feature in pandas.

```python
dropped_gun["participant_age"]= dropped_gun["participant_age"].astype(str) #setting type as string
dropped_gun["participant_age"]= dropped_gun["participant_age"].str.slice(start = -2) #taking only the last 2, which represent the age
dropped_gun["participant_age"]= dropped_gun["participant_age"].str.replace(":", "") #get rid of any extra colons
```

We'll also go ahead and convert all these values to int, now that there are no non-int values in our formatted data table.

```python
dropped_gun["participant_age"] = dropped_gun["participant_age"].astype(str).astype(int)
```

```python
gun_violence.head(10)
```

There are no NAN values here, and the values for age are formatted excellentely. We can also go ahead and rename this column for organizational purposes.

```python
dropped_gun = dropped_gun.rename(columns={"participant_age": "Age of Victim"})
```

We can also use imputing to clean up by inserting the average age of the non-null values into the null values.



```python
replaced_gun = gun_violence.fillna(":::" + str(dropped_gun["Age of Victim"].mean())) ##we're adding extra colons here so that
replaced_gun["participant_age"]= replaced_gun["participant_age"].astype(str) #setting type as string
replaced_gun["Age of Victim"]= replaced_gun["participant_age"].str.slice(start = -2) #taking only the last 2, which represent the age
replaced_gun["participant_age"]= replaced_gun["participant_age"].str.replace(":", "0") #get rid of any extra colons
```

# Simple Solution

We can simply take the mean age of all gun violence victims in this data set in order to determine at what the most likely age for being a victim of a gun violence crime is.

```python
dropped_gun["Age of Victim"].mean()
```

From this simple solution, the average age is for being a victim of gun violence in the United States is 29.7 years old.

#Statistical Test

Let us dig further into our distribution histogram. We can first use the Shapiro-Wilk Test to determine if our distribution is Gaussian or not. This function essentially calculates the P value and determines if the p value is below 0.05 or not. If so, our graph is gaussian. Otherwise, it is skewed.

```python
from scipy.stats import norm
from scipy.stats import shapiro
dropped_gun["Age of Victim"] = norm.rvs(size=2920)
shapiro(dropped_gun["Age of Victim"])
```

This is not a gaussian distribution since the p-value is FAR greater than 0.05, likely because the data is skewed towards the right, which is not representative of a normal distirbution.

From an initial observation of the visualization, we can conclude that a majority of victims of gun violence are in their 20s. This makes sense, as the data is skewed left as proven by the Shapiro-Wilk test that we conducted earlier. Additionally, according to the histogram, a plurality of American gun violence victims are of age 23, since they had the most values on the histogram plot. The histogram also has counts above one hundred, which is a lot in this sense once we cross around twenty years old. It begins to subside around thirty years old. As such, we can conclude than an American in his mid twenties is  most likely to be a victim of gun violence.
