**MAKING A PLOT:** We can use a line plot here, with three different lines for each accident. There are three different x-axises here for each type of accident and an y-axis representing the insurance premiums. 

The graph in this case will determine if as the percentage of accidents increases, will the premiums increase? And if so, which one increases at the highest rate?

Let's start by importing matplotlib to graph our data

```python
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
# This next line tells jupyter to render the images inline
%matplotlib inline
```

We'll sort the the data so that the amount of accidents is ordered.:

```python
bad_drivers = bad_drivers.sort_values(by=['Percentage Of Drivers Involved In Fatal Collisions Who Were Speeding'])
```

Now we can plot our data

```python
bad_drivers.plot.scatter("Percentage Of Drivers Involved In Fatal Collisions Who Were Speeding", "Car Insurance Premiums ($)")
```

```python
bad_drivers = bad_drivers.sort_values(by=['Percentage Of Drivers Involved In Fatal Collisions Who Were Alcohol-Impaired'])
bad_drivers.plot.scatter("Percentage Of Drivers Involved In Fatal Collisions Who Were Alcohol-Impaired", "Car Insurance Premiums ($)")
```

```python
bad_drivers = bad_drivers.sort_values(by=['Percentage Of Drivers Involved In Fatal Collisions Who Were Not Distracted'], ascending=False)
bad_drivers.plot.scatter("Percentage Of Drivers Involved In Fatal Collisions Who Were Not Distracted", "Car Insurance Premiums ($)")
```

Upon examining the plots, we can see that certain types of accidents have an effect on insurance prices. Upon examining the graph depicting states' percentage of accidents involving speeding versus car insurance premiums, we notice the graph is volatile across its entire domain. As such, there is likely no effect of accidents related to speeding on insurance prices

DWI related accidents, on the otherhand, appear to start creating price surges in certain states, which were going above the average price of 866 dollars. Once the percentage of driver's involved in DWI related accidents crosses around 23%, the graph begins to fluctuate. This likely means that insurance companies across different states start to make decisions or not to increase their prices in different states. 

Lastly, distraction related accidents appear to matter the least. The less accidents that involve distraction, the more the premiums start to fluctutate, with most of the samples rising above the average value of 866 dollars.

Once again, the data appeared to be intact; from just looking at the head and tail, it appeared that the speeding accidents had no effect on insurance costs. For my visualization, I chose a scatter plot. This is because I want to examine individual data points, which in this case are states. Using a line plot, in this case, would imply that the values between individual data points hold significance, when in reality, they do not. Using a scatter plot would also allow me to see where a majority of data points are trending. Upon plotting the percentage of drivers involved in fatal collisions who were speeding versus insurance premiums, the plot showed to be very volatile, and had no conclusive result. It can be safe to conclude from this data, as such, that speeding related accidents had no effect on insurance premiums. For DWI related accidents, the points began to trend upward once the percentage crossed around 25%. A majority of points were above the mean value of 866 dollars. Finally, I examine the plot for distraction related accidents. This data was unusually labeled, as the x-axis given in the CSV file was the percentage of drivers NOT involved in distraction related accidents. As such, I read the graph backwards. The higher the percentage, the less percentage of drivers that were involved in distraction related accidents. As the percentage of drivers involved in distraction related accidents decreased, the points began to fluctuate, and more states had higher insurance premiums. This does not mean that insurance companies charge higher, but rather they tend to charge higher for other types of accidents since the value is a percentage.

We can therefore conclude from this analysis that insurance companies tend to charge higher in states with high DWI accidents compared to other types of accidents.
